### Solución Estudiante 1

```python
# Código de solución estudiante 1
import pandas as pd
import numpy as np
import scipy.stats as st     # Librería estadística
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(index=['Pico','pbAbajo','pbSubiendo', 'pbBajando', 'total_ventas'],columns=['Alta','Mediana','Baja','total_ec'])
data.iloc[0,:-1] = [20,7,3]
data.iloc[1,:-1] = [30,40,30]
data.iloc[2,:-1] = [20,8,2]
data.iloc[3,:-1] = [30,5,5]
data['total_ec'] = data.sum(axis=1)
data.loc['total_ventas',:] = data.sum(axis=0)
data
```

```python
Pa = data.iloc[4,0]/data.iloc[-1,-1]
Pm = data.iloc[4,1]/data.iloc[-1,-1]
Pb = data.iloc[4,2]/data.iloc[-1,-1]
P = np.array([Pa, Pm, Pb])
f_o = data.loc[data.index!='total_ventas',data.columns!='total_ec'].values.flatten()
f_e = np.concatenate([data.iloc[0,-1]*P, data.iloc[1,-1]*P, data.iloc[2,-1]*P, data.iloc[3,-1]*P])
f_o, f_e
```

```python
#### Método chi cuadrado
F_obse = f_o #[20, 7, 3, 30, 40, 30, 20, 8, 2, 30, 5, 5]
F_espe = f_e #[15.,  9.,  6., 50., 30., 20., 15.,  9.,  6., 20., 12.,  8.]
GDL = (4-1)*(5-1)
x2 = st.chisquare(F_obse,F_espe,ddof=GDL)
print('Valor de chi cuadrado = ',list(x2)[0],',p-value de la prueba=',list(x2)[1])

Ji = st.chi2.ppf(q = 0.9,df=GDL)
print('Estadístico de Ji = ',Ji)
# st.t.interval
```

```python
x1 = st.chi2.cdf(list(x2)[0],df=GDL)
x1
```
