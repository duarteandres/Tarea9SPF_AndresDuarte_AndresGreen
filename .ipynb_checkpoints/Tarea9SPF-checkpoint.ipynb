### Solución Estudiante 1

```python
# Código de solución estudiante 1
import pandas as pd
import numpy as np
import scipy.stats as st     # Librería estadística
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(index=['Pico','pbAbajo','pbSubiendo', 'pbBajando', 'total_ventas'],columns=['Alta','Mediana','Baja','total_ec'])
data.iloc[0,:-1] = [20,7,3]
data.iloc[1,:-1] = [30,40,30]
data.iloc[2,:-1] = [20,8,2]
data.iloc[3,:-1] = [30,5,5]
data['total_ec'] = data.sum(axis=1)
data.loc['total_ventas',:] = data.sum(axis=0)
data
```

```python
Pa = data.iloc[4,0]/data.iloc[-1,-1]
Pm = data.iloc[4,1]/data.iloc[-1,-1]
Pb = data.iloc[4,2]/data.iloc[-1,-1]
P = np.array([Pa, Pm, Pb])
f_o = data.loc[data.index!='total_ventas',data.columns!='total_ec'].values.flatten()
f_e = np.concatenate([data.iloc[0,-1]*P, data.iloc[1,-1]*P, data.iloc[2,-1]*P, data.iloc[3,-1]*P])
f_o, f_e
```

```python
#### Método chi cuadrado
F_obse = f_o #[20, 7, 3, 30, 40, 30, 20, 8, 2, 30, 5, 5]
F_espe = f_e #[15.,  9.,  6., 50., 30., 20., 15.,  9.,  6., 20., 12.,  8.]
GDL = (3-1)*(4-1)
x2 = st.chisquare(F_obse,F_espe,ddof=GDL)
print('Valor de chi cuadrado = ',list(x2)[0],',p-value de la prueba=',list(x2)[1])

Ji = st.chi2.ppf(q = 0.9,df=GDL)
print('Estadístico de Ji = ',Ji)
# st.t.interval
```

```python
x1 = st.chi2.cdf(list(x2)[0],df=GDL)
x1
```

## Respuesta planteada en el ejercicio 1
### Rechazamos que el nivel de las ventas sea independiente del ciclo de negocios de la economía del país, ya que el valor  de $Xi^2$ quedó más allá de nuestro valor crítico.

Ejercicio 2

```python
data2 = pd.DataFrame(index=['rookies','vet_notStars','vet_Stars'],columns=['Prob', 'Muestra'])
data2['Prob'] = [0.30,0.60,0.10]
data2['Muestra'] = [50, 45, 5]
data2
```

```python
GDL2 = 2

f_o2 = np.array(data2['Muestra'], dtype=object)
f_e2 = np.array(data2['Prob']*100, dtype=object)


f_o2, f_e2
```

```python
x2_2 = st.chisquare(f_o2, f_e2, ddof=2-.0001)
print('Valor de chi cuadrado = ',list(x2_2)[0],', p-value de la prueba=',list(x2_2)[1])

Ji_2 = st.chi2.ppf(q = 0.95, df=2)
print('Estadístico de Ji = ',Ji_2)
```

## Respuesta planteada en el ejercicio 2
### Rechazamos $H_0$ porque el p-value es menor a .05, la proporción de las cartas no es la mencionada (30, 60 y 10% respectivamente)

Ejercicio 3

```python
data3 = pd.DataFrame(index=['Boys','Girls','total2'],columns=['Lone Ranger','Sesame Street','The Simpsons','total1'])
data3.iloc[0,:-1] = [50,30,20]
data3.iloc[1,:-1] = [50,80,70]
data3['total1'] = data3.sum(axis=1)
data3.loc['total2',:] = data3.sum(axis=0)
data3
```

```python
Ptv = data3.iloc[0,3]/data3.iloc[-1,-1]
f_o3 = data3.loc[data3.index!='total2',data3.columns!='total1'].values.flatten()
f_e3 = np.concatenate([data3.iloc[-1,:-1]*Ptv,data3.iloc[-1,:-1]*(1-Ptv)])

f_e3, f_o3, Ptv
```

```python
GDL3 = 2

x2_3 = st.chisquare(f_o3, f_e3, ddof=GDL3)
print('Valor de chi cuadrado = ', list(x2_3)[0],',p-value de la prueba=',list(x2_3)[1])

Ji_3 = st.chi2.ppf(q = 0.95, df=GDL3)
print('Estadístico de Ji = ',Ji_3)
```

## Respuesta planteada en el ejercicio 3
### Rechazamos $H_0$ porque el p-value es menor a .05, las preferencias de los niños sí difieren de las de las niñas en cuanto a televisión.
